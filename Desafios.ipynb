```python
diametro_naranja = datos[:5000, 0]
diametro_toronja = datos[5000:, 0]
peso_naranja = datos[:5000, 1]
peso_toronja = datos[5000:, 1]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(peso_naranja, diametro_naranja)
plt.plot(peso_toronja, diametro_toronja)
plt.legend(["Naranja","Toronja"])
```

$$ \alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2}$$

```python
# Naranjas
n = np.size(peso_naranja)
y = peso_naranja
x = diametro_naranja
# Toronjas
n_T = np.size(peso_toronja)
y_T = peso_toronja
x_T = diametro_toronja
```

```python
a = (n * np.sum(x*y) - np.sum(x) * np.sum(y))/ (n *   np.sum(x**2)- np.sum(x)**2)
a_T = (n_T * np.sum(x_T*y_T) - np.sum(x_T) * np.sum(y_T))/ (n_T *   np.sum(x_T**2)- np.sum(x_T)**2)
```

$$ \beta = \frac{\sum y - \alpha \sum x}{n}$$

```python
b = (np.sum(y) - a * np.sum(x))/n
b_T = (np.sum(y_T) - a_T * np.sum(x_T))/n_T
```

```python
y = a*x + b
y_T =a_T * x_T + b_T
```

```python
plt.plot(diametro_naranja, y)
plt.plot(diametro_naranja, peso_naranja)
plt.legend(["Regresión lineal","Peso naranja"])
```
