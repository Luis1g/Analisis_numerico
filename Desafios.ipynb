```python
import numpy as np
url = "https://gist.githubusercontent.com/ahcamachod/9be09de793dc3bf1e6c3d98eb4e5b1ef/raw/21b85572693200040e11284ef6dcfc3457ec8e11/citrus.csv"
datos = np.loadtxt(url, delimiter=",", usecols=np.arange(1, 6, 1), skiprows=1)
```

Ha llegado el momento de poner a prueba los conocimientos adquiridos durante la lección 2. Continuando con el proyecto de las naranjas/toronjas, ahora debes seleccionar parte de los datos. Las columnas que evaluaremos son el diámetro y el peso. Crea arrays específicos para almacenar el diámetro y el peso de la naranja y la toronja. El diámetro está en la columna cero y el peso en la columna uno. Los datos de las naranjas van hasta la fila 4999 y los datos de las toronjas comienzan en la fila 5000 del archivo.

Después de seleccionar los datos, importa la biblioteca matplotlib y crea un gráfico para el peso en función del diámetro tanto para las naranjas como para las toronjas.

```python
diametro_naranja = datos[:5000, 0]
diametro_toronja = datos[5000:, 0]
peso_naranja = datos[:5000, 1]
peso_toronja = datos[5000:, 1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(peso_naranja, diametro_naranja)
plt.plot(peso_toronja, diametro_toronja)
plt.legend(["Naranja","Toronja"])
```
